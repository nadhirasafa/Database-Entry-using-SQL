## 1.   Preparation and Connect to Sqlite3







```python
import sqlite3
import pandas as pd
```

```python
db = sqlite3.connect('/content/drive/MyDrive/Karier.mu/iowa_sales.db')
```

## 2.   Check Tables and Drop Unrelated Table



```python
cursor = db.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
pd.read_sql_query("SELECT*from employee", db)
```

Setelah dilihat isi tabelnya, tabel employee dihapus karena kurang berkaitan dengan database yang berisi tentang penjualan minuman asal Amerika. Data karyawan yang tampil pun bukan karyawan yang bekerja langsung di bagian penjualan. Selain itu, sebagai relational database, tabel employee tidak memiliki kolom/atribut yang dapat menghubungkan dengan tabel-tabel lainnya yang akan diinput setelah ini (tidak ada yang bisa jadi foreign key) sehingga  dapat disimpulkan rendahnya korelasi. Oleh karena itu, tabel tersebut akan dihapus.

```python
cursor_obj=db.cursor()
cursor_obj.execute("drop table employee")
```

```python
#cek apakah tabel employee sudah terhapus
cursor_obj.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor_obj.fetchall())
```

## 4. Import Excel File to SQLite3

```python
table2="""CREATE TABLE sales (
  store_number INT(30) NOT NULL,
  store_name VARCHAR(500) NOT NULL,
  city VARCHAR(150) NOT NULL,
  county_number INT(20) NOT NULL,
  county VARCHAR(150) NOT NULL,
  category_name VARCHAR(300) NOT NULL,
  item_number INT(30) NOT NULL,
  item_description VARCHAR(200) NOT NULL,
  pack INT(10) NOT NULL,
  bottle_volume_ml INT(10) NOT NULL,
  state_bottle_cost FLOAT(20) NOT NULL,
  state_bottle_retail FLOAT(20) NOT NULL,
  bottles_sold INT(20) NOT NULL,
  sale_dollars FLOAT(30) NOT NULL
);"""

try:
  cursor_obj.execute(table2)
  print("Table is Ready.")
except Exception as e:
  print("There is something wrong!")
  print(e)
```

```python
#cek apakah tabel sales sudah muncul
cursor_obj.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor_obj.fetchall())
```

```python
#melihat struktur table beserta kolom2nya
structure = """ PRAGMA table_info(sales); """
pd.read_sql_query(structure, db)
```

## e. Entry data yang memiliki nilai sale_dollars kurang dari 100 dan lebih dari 1000 dihapus

```python
#melihat data yang belum diganti
pd.read_sql_query("SELECT sale_dollars from SALES where NOT (sale_dollars>100 AND sale_dollars<1000) ORDER BY sale_dollars",db)
```
